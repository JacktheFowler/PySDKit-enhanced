```python
import numpy as np
from pysdkit import EMD
from pysdkit.data import test_emd, plot_signal
from pysdkit.plot import plot_IMFs
```

```python
# create an example signal
t, signal = test_emd()
# print(signal.shape)
fig = plot_signal(t, signal)
```

```python
# create an instance for signal decomposition
emd = EMD()
IMFs = emd.fit_transform(signal, max_imfs=2)
plot_IMFs(signal, IMFs)
```

```python
from pysdkit import EEMD

eemd = EEMD(ext_EMD=emd)
IMFs = eemd.fit_transform(signal, max_imfs=2)
print(IMFs.shape)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import CEEMDAN

ceemdan = CEEMDAN(ext_EMD=emd)
IMFs = ceemdan.fit_transform(signal, max_imfs=2)
print(IMFs.shape)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import EWT

ewt = EWT(K=3)
IMFs = ewt.fit_transform(signal)
print(IMFs.shape)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import ACMD

acmd = ACMD(K=3)
IFMs = acmd.fit_transform(signal)
print(IMFs.shape)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import LMD

lmd = LMD(K=1)
IMFs = lmd.fit_transform(signal)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import SSA

ssa = SSA(K=2)

IMFs = ssa.fit_transform(signal)

plot_IMFs(signal, IMFs)
```

```python
from pysdkit import VMD

# load new signal
signal = np.load("example.npy")

# use variational mode decomposition
vmd = VMD(alpha=500, K=3, tau=0.0, tol=1e-9)
IMFs = vmd.fit_transform(signal=signal)
print(IMFs.shape)

vmd.plot_IMFs(save_figure=True, dpi=600)
```

```python
from pysdkit.plot import plot_IMFs_amplitude_spectra

# frequency domain visualization
plot_IMFs_amplitude_spectra(IMFs, smooth="exp")   # use exp smooth
```
